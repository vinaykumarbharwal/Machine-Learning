# 1.OneHotEncoding using Pandas

```python
df_encoded = pd.get_dummies(df, columns=['fuel', 'owner'])

# Convert only columns that start with 'fuel_' or 'owner_'
for col in df_encoded.columns:
    if col.startswith('fuel_') or col.startswith('owner_'):
        df_encoded[col] = df_encoded[col].astype(int)
df_encoded
```

# 2. K-1 OneHotEncoding

```python
df_encoded = pd.get_dummies(df, columns=['fuel','owner'], drop_first=True)
df_encoded[df_encoded.columns.difference(df.columns)] = df_encoded[df_encoded.columns.difference(df.columns)].astype(int)
df_encoded
```
