```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

final1 = pd.DataFrame()

for j in range(0, 501):
    url = f'https://www.ambitionbox.com/list-of-companies?page={j}'
    
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/121.0.0.0 Safari/537.36",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
        "Accept-Language": "en-US,en;q=0.9",
        "Referer": "https://www.ambitionbox.com/",
        "Connection": "keep-alive",
    }
    
    webpage = requests.get(url, headers=headers).text
    
    try:
        soup = BeautifulSoup(webpage, "lxml")
    except:
        soup = BeautifulSoup(webpage, "html.parser")
    
    company_cards = soup.find_all('div', class_='companyCardWrapper')
    
    name = []
    rating = []
    reviews = []
    company_type = []
    location_info = []

    for i in company_cards:  # changed card -> i
        name.append(i.find('h2').text.strip())
        rating_tag = i.find('div', class_='rating_text')
        rating.append(rating_tag.text.strip() if rating_tag else None)
        
        reviews_tag = i.find('span', class_='companyCardWrapper__companyRatingCount')
        reviews.append(reviews_tag.text.strip() if reviews_tag else None)
        
        span = i.find('span', class_='companyCardWrapper__interLinking')
        if span and '|' in span.text:
            ind, loc = [p.strip() for p in span.get_text(strip=True).split('|', 1)]
        else:
            ind, loc = None, None

        company_type.append(ind)
        location_info.append(loc)
    
    # Create DataFrame for this page and append to final1
    df = pd.DataFrame({
        "Company_Name": name,
        "Rating": rating,
        "Reviews": reviews,
        "Company_Type": company_type,
        "Location_Info": location_info
    })
    
    final1 = pd.concat([final1, df], ignore_index=True)

print(final1)
```

```python
final1.to_csv('ambition_box_web_scraping.csv')
```
