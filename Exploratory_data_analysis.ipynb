```python
import pandas as pd

df = pd.read_csv("stack-overflow-developer-survey-2025/survey_results_public.csv",low_memory=False)
```

```python
df.shape
df.columns
```

```python
drop_patterns = [
    'ResponseId',
    'ConvertedCompYearly'
]

drop_suffixes = ['_TEXT', 'HaveEntry', 'WantEntry']

cols_to_drop = [
    col for col in df.columns
    if any(col.endswith(suf) for suf in drop_suffixes)
    or col in drop_patterns
]

df_clean = df.drop(columns=cols_to_drop, errors='ignore')
```

```python
single_questions = [
    'MainBranch','Age','EdLevel','Employment','EmploymentAddl',
    'WorkExp','YearsCode','DevType','OrgSize','ICorPM','RemoteWork',
    'Industry','Country','Currency','CompTotal','JobSat','AIThreat',
    'NewRole','SOAccount','SOVisitFreq','SODuration','SOPartFreq',
    'SOComm','SOFriction','AISelect','AISent','AIAcc','AIComplex',
    'AIFrustration','AIExplain','AIAgents','AIAgentChange','AIHuman','AIOpen'
]
```

```python
multi_questions = {
    'LanguageHaveWorkedWith': 'Languages',
    'LanguageWantToWorkWith': 'Languages',
    'LanguageAdmired': 'Languages',
    'DatabaseHaveWorkedWith': 'Databases',
    'DatabaseWantToWorkWith': 'Databases',
    'DatabaseAdmired': 'Databases',
    'PlatformHaveWorkedWith': 'Platforms',
    'PlatformWantToWorkWith': 'Platforms',
    'PlatformAdmired': 'Platforms',
    'WebframeHaveWorkedWith': 'Web Frameworks',
    'WebframeWantToWorkWith': 'Web Frameworks',
    'WebframeAdmired': 'Web Frameworks',
    'DevEnvsHaveWorkedWith': 'Dev Environments',
    'DevEnvsWantToWorkWith': 'Dev Environments',
    'DevEnvsAdmired': 'Dev Environments',
    'AIModelsHaveWorkedWith': 'AI Models',
    'AIModelsWantToWorkWith': 'AI Models',
    'AIModelsAdmired': 'AI Models'
}
```

```python
matrix_questions = {
    'JobSatPoints': 'Job Satisfaction',
    'TechEndorse': 'Technology Attitudes',
    'TechOppose': 'Technology Attitudes',
    'SO_Actions': 'Stack Overflow',
    'AIAgentImpact': 'AI Agents',
    'AIAgentChallenges': 'AI Agents',
    'AITool': 'AI Tools'
}
```

```python
schema_rows = []
qid_counter = 1
```

```python
for col in single_questions:
    if col in df_clean.columns:
        schema_rows.append({
            'qid': f'Q{qid_counter:03}',
            'qname': col,
            'question': col,
            'type': 'single',
            'sub': col,
            'sq_id': col
        })
        qid_counter += 1
```

```python
for col, sub in multi_questions.items():
    if col in df_clean.columns:
        schema_rows.append({
            'qid': f'Q{qid_counter:03}',
            'qname': col,
            'question': col,
            'type': 'multi',
            'sub': col,
            'sq_id': col
        })
        qid_counter += 1
```

```python
for prefix, sub in matrix_questions.items():
    matrix_cols = [c for c in df_clean.columns if c.startswith(prefix)]
    
    for c in matrix_cols:
        sq = c.split('_')[-1] if '_' in c else None
        
        schema_rows.append({
            'qid': f'Q{qid_counter:03}',
            'qname': prefix,
            'question': prefix,
            'type': 'matrix',
            'sub': col,
            'sq_id': col
        })
    qid_counter += 1
```

```python
schema_df = pd.DataFrame(schema_rows)
```

```python
schema_df['question'] = schema_df['question'].str.replace('  ', ' ')
schema_df['qname'] = schema_df['qname'].str.strip()
```

```python
schema_df = schema_df.sort_values(['qid','sq_id'])
```

```python
schema_df.to_csv("survey_results_schema.csv", index=False)
```

```python
import matplotlib.pyplot as plt

# 1. Calculate the total number of respondents for percentage calculation
total_respondents = df['LanguageHaveWorkedWith'].dropna().shape[0]

# 2. Process data as before
lang_series = df['LanguageHaveWorkedWith'].dropna().str.split(';')
lang_counts = lang_series.explode().value_counts().sort_values(ascending=True)

# 3. Create the plot with a larger size (width, height)
plt.figure(figsize=(12, 8)) 
ax = lang_counts.plot(kind='barh', color='skyblue', edgecolor='black')

# 4. Add percentage labels next to each bar
for i, value in enumerate(lang_counts):
    # Calculate percentage based on total unique respondents
    percent = (value / total_respondents) * 100
    
    # Place text: x = value + a small offset, y = index (i)
    plt.text(value + 0.1, i, f'{percent:.1f}%', va='center', fontsize=11, fontweight='bold')

# 5. Formatting for better readability
plt.title('Programming Languages Worked With', fontsize=16, pad=20)
plt.xlabel('Number of Respondents', fontsize=12)
plt.ylabel('Language', fontsize=12)

# Adjust the x-axis limit so labels don't get cut off
plt.xlim(0, max(lang_counts) * 1.15) 

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#####################################################################
import pandas as pd

public_df = pd.read_csv("stack-overflow-developer-survey-2025/survey_results_public.csv",low_memory=False)
schema_df = pd.read_csv("stack-overflow-developer-survey-2025/survey_results_schema.csv")
```

```python
public_df.info()
schema_df.head()
```

```python
drop_patterns = [
    "_TEXT",
    "Entry",
    "Choice"
]

cols_to_drop = [
    col for col in public_df.columns
    if any(p in col for p in drop_patterns)
]

public_df = public_df.drop(columns=cols_to_drop, errors="ignore")
```

```python
public_df.columns = (
    public_df.columns
    .str.strip()
    .str.replace(" ", "_", regex=False)
)
```

```python
numeric_cols = [
    "WorkExp",
    "YearsCode",
    "CompTotal",
    "ConvertedCompYearly",
    "ToolCountWork",
    "ToolCountPersonal"
]

for col in numeric_cols:
    public_df[col] = pd.to_numeric(public_df[col], errors="coerce")
```

```python
public_df["YearsCode"] = public_df["YearsCode"].replace({
    "Less than 1 year": 0,
    "More than 50 years": 50
})
```

```python
cat_cols= public_df.select_dtypes(include="object").columns

public_df[cat_cols] = public_df[cat_cols].apply(
    lambda x: x.str.strip()
)
```

```python
matrix_cols = (
    public_df.filter(like="JobSatPoints").columns.tolist() +
    public_df.filter(like="TechEndorse_").columns.tolist() +
    public_df.filter(like="TechOppose_").columns.tolist() +
    public_df.filter(like="SO_Actions_").columns.tolist()
)

public_df[matrix_cols] = public_df[matrix_cols].apply(
    pd.to_numeric, errors="coerce"
)
```

```python
multi_select_cols = [
    "LanguageHaveWorkedWith",
    "LanguageWantToWorkWith",
    "DatabaseHaveWorkedWith",
    "PlatformHaveWorkedWith",
    "WebframeHaveWorkedWith",
    "DevEnvsHaveWorkedWith",
    "AIModelsHaveWorkedWith"
]

for col in multi_select_cols:
    public_df[col] = (
        public_df[col]
        .str.lower()
        .str.split(";")
    )
```

```python
final_columns = [
    "ResponseId",   # ← keep this
    "MainBranch","Age","EdLevel","Employment","WorkExp","YearsCode",
    "DevType","OrgSize","ICorPM","RemoteWork","Industry","Country",
    "Currency","CompTotal","ConvertedCompYearly","JobSat","AIThreat",
    "AISent","AIAcc","AIComplex","AIFrustration","AIAgents",
    "LanguageHaveWorkedWith","DatabaseHaveWorkedWith",
    "PlatformHaveWorkedWith","AIModelsHaveWorkedWith"
]

clean_df = public_df[final_columns]
```

```python
long_df = clean_df.melt(
    id_vars=["ResponseId"],
    var_name="qname",
    value_name="response"
)
final_df = long_df.merge(
    schema_df,
    on="qname",
    how="left"
)
```
