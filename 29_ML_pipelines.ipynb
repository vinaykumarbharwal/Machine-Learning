```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
# Step 1 -> train/test/split 
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']),
                                                    df['Survived'], 
                                                    test_size=0.2, 
                                                    random_state=42)
```

```python
X_train.head(5)
```

```python
y_train.sample(5)
```

```python
df.isnull().sum()
```

```python
# Applying imputation

si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

```python
X_train_age
```

```python
X_train_embarked
```

```python
# one hot encoding Sex and Embarked
ohe_sex = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked)
```

```python
X_train_sex
```

```python
X_train_embarked
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked']) 
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_sex,X_train_embarked),axis=1) 
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_sex,X_test_embarked),axis=1)
```

```python
X_train_transformed
```

```python
X_test_transformed
```

```python
clf = DecisionTreeClassifier() 
clf.fit(X_train_transformed,y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)
```

# 2. Predict using Pipeline

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('titanic-dataset.csv')
```

```python
df.head()
```

## Lets plan

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# Step 1 -> train/test/split 
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']),
                                                    df['Survived'], 
                                                    test_size=0.2, 
                                                    random_state=42)
```

```python
X_train.head()
```

```python
y_train.sample(5)
```

```python
# imputation transformer
trf1 = ColumnTransformer([
    ('impute_age', SimpleImputer(), [2]),
    ('impute_embarked', SimpleImputer(strategy='most_frequent'), [6])
], remainder='passthrough')
```

```python
# one hot encoding
trf2 = ColumnTransformer([
    ('ohe_sex_embarked', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [1,6])
], remainder='passthrough')
```

```python
#scaling
trf3=ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
#Feature selection
trf4=SelectKBest(score_func=chi2,k=8)
```

```python
#train the model
trf5=DecisionTreeClassifier()
```

# Create pipeline

```python
pipe = Pipeline([
    ('trf1', trf1),  # Imputation transformer
    ('trf2', trf2),  # One-hot encoding transformer
    ('trf3', trf3),  # Likely scaling or feature selection
    ('trf4', trf4),  # Possibly model-specific preprocessing
    ('trf5', trf5)   # Final estimator (e.g., classifier)
])
```

# Pipeline Vs make_pipeline  
### Pipeline requires naming of steps, make_pipeline does not.  
### Same applies to ColumnTransformer vs make_column_transformer


```python
#alternate syntax
pipe1=make_pipeline(trf1,trf2,trf3,trf4,trf5)
```

```python
pipe.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Cross Validation using Pipeline

```python
# cross validation using cross_val_score
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,X_train,y_train,cv=5,scoring='accuracy').mean()
```

# GridSearch using Pipeline

```python
# Define hyperparameter grid for the final estimator (trf5)
params = {
    'trf5__max_depth': [1, 2, 3, 4, 5, None]  # Try different tree depths including no limit
}
```

```python
# Import GridSearchCV for hyperparameter tuning
from sklearn.model_selection import GridSearchCV

# Setup grid search with 5-fold cross-validation and accuracy scoring
grid = GridSearchCV(pipe, params, cv=5, scoring='accuracy')

# Fit the grid search on training data
grid.fit(X_train, y_train)
```

```python
# Display the best cross-validated score
grid.best_score_
```
