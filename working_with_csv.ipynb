***13. Handling dates***

```python
pd.read_csv('italy-covid-daywise.csv',parse_dates={'date'})
```

```python
def rename(name):
    if name=="Asia":
        return "AS"
    else:
        return name
```

```python
rename("Asia")
```

***14. convertors***

```python
pd.read_csv('locations.csv',converters={'continent':rename})
```

***15. na_values parameters***

```python
pd.read_csv('locations.csv',na_values=['Africa'])
```
